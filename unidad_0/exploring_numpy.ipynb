## Numpy

NumPy es fundamental en Python para computación científica y análisis de datos. Sus funciones clave permiten crear, manipular y operar eficientemente con arrays n-dimensionales (ndarrays)

Partimos por importar numpy

```python
import numpy as np
```

### Arrays

Creamos una lista y la transformamos en array. Estos son estructuras de datos homogéneas, multidimensionales y de tamaño fijo.

```python
np.array([1, 2, 3, 4, 5])
```

Si trato de crear una con otro tipo de dato, por ejemplo, agregar una string, numpy transforma todo el array a dtype string.

```python
np.array([1, 2, 3, 4, "gato"])
```

Si introduzco un número decimal, transforma todo el array en float.

```python
np.array([1.1, 2, 3.4, 4, 5])
```

Podemos generar algunos arrays o matrices definiendo sus tamaños. 

Por ejemplo, una matriz de 1s o 0s.

```python
np.ones(100)
```

```python
np.zeros(100)
```

Un array que cubra un espacio homogeneamente.

```python
np.linspace(0, 10, 101) #En linspace, el último número es el número de elementos.
```

O por otro lado, un array que tenga elementos x-espaciados.

```python
np.arange(0, 5, 0.1) #En arange, el último número es tamaño del paso.
```

Podemos generar una matriz de números aleatorios

```python
data = np.random.randn(5, 4) #La matriz tiene 5 filas y 5 columnas.
data
```

Podemos cambiar el la forma de la matriz

```python
data.reshape(2, 10) #Ahora la matriz tiene 2 filas y 10 columnas.
```

Se pueden concatenar matrices y arrays en sus diferentes ejes

```python
np.concatenate((data, data), axis = 0)
```

```python
np.concatenate((data, data), axis = 1)
```

Y podemos aplanar la matriz para generar un array de una dimensión

```python
data.flatten() #Nos queda un array de una dimensión y 25 elementos.
```

### Estadísticas
Podemos hacer un poco de estadística descriptiva utilizando funciones de numpy

Por ejemplo, concocer el valor de la suma de nuestros datos

```python
np.sum(data)
```

Promedio, desviación estandar y percentiles

```python
np.mean(data, axis = 0)
```

```python
np.std(data, axis = 0)
```

```python
q1 = np.quantile(data, 0.25) #Cuartil 1
median = np.quantile(data, 0.5) #Mediana
q3 = np.quantile(data, 0.75) #Cuartil 3
display(q1, median, q3)
```

### Algebra lineal


NumPy es la librería fundamental para el álgebra lineal en Python. Permite obtener operaciones extensivas en cálculos de forma eficiente.

```python
matrix = np.array([[0.1, 0.5], [-0, -0.1]])
matrix
```

Por ejemplo, obtener la inversa de una matrix

```python
inverse = np.linalg.inv(matrix)
inverse
```

Realizar multiplicaciones matriciales (un proceso bastante costoso computacionalmente).

```python
matrix@inverse #Nos da la matriz identidad.
```

Cálcular determinantes.

```python
np.linalg.det(matrix)
```

Descomposiciones de matrices, obtención de vectores y valores propios:

```python
eigenvectors, eigenvalues = np.linalg.eig(matrix)
display(eigenvectors, eigenvalues)
```

O incluso Singular value decomposition.

```python
U, s, V = np.linalg.svd(matrix)
display(U, s, V)
```

Además, podemos utilizar numpy resolver sistemas de ecuaciones lineales.

```python
np.linalg.solve(matrix, np.array([1, 1]))
```
