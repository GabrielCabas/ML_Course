# Matplotlib

Matplotlib es una biblioteca de código abierto para Python diseñada para la visualización de datos, permitiendo crear gráficos 2D y 3D de alta calidad, estáticos, animados e interactivos.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

El modelo mental de matplotlib es el siguiente:

```
Figure
 └── Axes
      ├── Axis (x, y)
      ├── Artists (líneas, scatter, texto, patches, etc.)
```
Obtenemos las variables fig y ax.

````python
fig, ax = plt.subplots()
````

Podemos generar mas de un axis por figure para realizar figuras mas complejas.

````python
fig, ax = plt.subplots(2, 1) # Número de filas y columnas.
````

Podemos compartir el eje x o y entre los axis.

````python
fig, ax = plt.subplots(2, 2, sharex=True)
````

## Gráficos que debemos dominar

Scatterplot

````python
x = np.random.randn(20)
y = np.random.randn(20)
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
````

Line plot

````python
x = np.linspace(0, 10, 20)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
````

Boxplots

````python
x = np.random.randn(20)
y = np.random.randn(20)
fig, ax = plt.subplots()
ax.boxplot([x, y])
plt.show()
````

Histogramas

````python
x = np.random.randn(500)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
````

Gráficos de barras

````python
y = np.random.randn(10)
x = np.arange(10)
fig, ax = plt.subplots()
ax.bar(x, y)
plt.show()
````

Heatmaps

````python
z = np.random.randn(10, 10)
fig, ax = plt.subplots()
ax.imshow(z)
plt.show()
````

## Personalización

La mayor fortaleza de matplotlib es la posibilidad de personalizar nuestros gráficos a detalle, cambiar el tipo de linea, colores, transparencia, textos, etc. 

A continuación mostramos un gráfico con algunas funciones matemáticas. Siéntete libre de cambiar lo que quieras :D 

````python
x = np.linspace(-2, 2, 100)
fig, ax = plt.subplots(2,2, sharey=True, sharex=True , constrained_layout=True)

#Funciones trigonométricas

sin = np.sin(x)
cos = np.cos(x)

ax[0,0].grid(True)
ax[0,0].set_xlabel("x")
ax[0,0].set_ylabel("y")

sin_line, = ax[0,0].plot(x, sin)
sin_line.set_color("salmon")
sin_line.set_label("sin(x)")

cos_line, = ax[0,0].plot(x, cos)
cos_line.set_color("steelblue")
cos_line.set_linestyle("--")
cos_line.set_label("cos(x)")

ax[0,0].legend()

ax[0,0].set_title("Trigonométricas")

#Funciones trigonométricas inversas
arcsin = np.arcsin(x)
arccos = np.arccos(x)


ax[0,1].grid(True)
ax[0,1].set_xlabel("x")
ax[0,1].set_ylabel("y")

arcsin_line, = ax[0,1].plot(x, arcsin)
arcsin_line.set_color("salmon")
arcsin_line.set_label("arcsin(x)")

arccos_line, = ax[0,1].plot(x, arccos)
arccos_line.set_color("steelblue")
arccos_line.set_label("arccos(x)")

ax[0,1].legend()
ax[0,1].set_title("Trigonométricas inversas")

#Funciones hiperbólicas
sinh = np.sinh(x)
cosh = np.cosh(x)
tanh = np.tanh(x)


ax[1,0].grid(True)
ax[1,0].set_xlabel("x")
ax[1,0].set_ylabel("y")

sinh_line, = ax[1,0].plot(x, sinh)
sinh_line.set_color("salmon")
sinh_line.set_label("sinh(x)")

cosh_line, = ax[1,0].plot(x, cosh)
cosh_line.set_color("steelblue")
cosh_line.set_label("cosh(x)")

tanh_line, = ax[1,0].plot(x, tanh)
tanh_line.set_color("green")
tanh_line.set_label("tanh(x)")
tanh_line.set_alpha(0.5)

ax[1,0].legend()
ax[1,0].set_title("Hiperbólicas")


#Funciones exponenciales
exp = np.exp(x)
log = np.log(x)

ax[1,1].grid(True, linestyle="--")
ax[1,1].set_xlabel("x")
ax[1,1].set_ylabel("y")

exp_line, = ax[1,1].plot(x, exp)
exp_line.set_color("salmon")
exp_line.set_label("exp(x)")

log_line, = ax[1,1].plot(x, log)
log_line.set_color("steelblue")
log_line.set_label("log(x)")
log_line.set_alpha(0.5)

ax[1,1].legend()
ax[1,1].set_title("Exponenciales")

````

## Seaborn

Seaborn es una capa de alto nivel construida sobre matplotlib, orientada a visualización estadística y basada en DataFrames (especialmente pandas).

````python
import seaborn as sns
````

Seaborn contiene algunos datasets de prueba que podemos usar en nuestras visualizaciones

````python
data = sns.load_dataset("iris")
data
````

Fíjate que seaborn entiende cuales son los nombres de las columnas de nuestro dataframe.


````python
ax = sns.scatterplot(x="sepal_length", y="sepal_width", data=data)
````

````python
ax = sns.boxplot(x="species", y="sepal_length", data=data)
````

La sintaxis es mucho mas simple, y además no es excluyente de matplotlib. Puedes utilizar el axis y seguir editando el gráfico

````python
ax = sns.histplot(data["sepal_length"])
ax.set_title("Histograma de Sepal Length")
ax.set_xlabel("Sepal Length")
````
