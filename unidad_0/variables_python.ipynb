## Tipos de variables en python

### Variables simples:

 - int: Números enteros.
 - float: Números decimales.
 - str: Cadenas de caracteres.
 - bool: Booleanos (True o False).


Python tiene tipado débil. Esto significa que las variables pueden tomar cualquier tipo de dato a lo largo del código.

```python
a = 10
b = 0.5
c = "Hola"
d = True
print("a es un número entero: ", a)
print("b es un número decimal: ", b)
print("c es una cadena de caracteres: ", c)
print("d es un booleano: ", d)
```

Podemos realizar algunas operaciones matemáticas y lógicas con estas variables.

```python
a + b
```

```python
a * b
```

```python
a * c
```

```python
d or not d
```

### Variables compuestas:

 - Tuplas.
 - Listas.
 - Sets.
 - Diccionarios.

```python
t = (1, 2, 3, 3, 4, 5, 5) #Tuplas
l  = [1, 2, 3, 3, "gato", 5, 5] #Listas
s = {1, 2, 3, 3, 4, 5, 5} #Sets
d = {"Gabriel": 26, "Susan": 34, "Evelin": 34} #Diccionarios
print("Tupla: ", t)
print("Lista: ", l)
print("Set: ", s)
print("Diccionario: ", d)
```

Las tuplas son datos inmutables: No se pueden agregar ni quitar elementos.

Podemos contar elementos repetidos

```python
print("El número 5 se repite", t.count(5), "veces. Está en la posición", t.index(5))
```

Las listas tienen un orden establecido.

```python
print("Este es el primer elemento de la lista a: ", l[0])
print("Este es el segundo elemento de la lista a: ", l[1])
print("Este es el tercer elemento de la lista a: ", l[2])
print("Este es el último elemento de la lista a: ", l[-1])
```

Podemos agregar elementos al final

```python
l.append(6)
print("Este es el último elemento de la lista a: ", l[-1]) #Utilizar - para invertir la búsqueda
print(l)
```

Podemos quitar elementos específicos

```python
l.pop() #Quitar el último elemento	
print("Este es el último elemento de la lista a: ", l[-1])
print(l)
l.pop(0) #Quitar el primer elemento
print("Este es el primer elemento de la lista a: ", l[0])
print(l)
```

Los sets no tienen elementos repetidos, tampoco un orden establecido.

```python
s
```

```python
try:
    print("Este es el primer elemento del set b: ", s[0])
except Exception as e:
    print(e, "Los sets no tienen un orden. Son simplemente 'conjuntos'.", sep="\n")
```

Los diccionarios son sets con una estructura key: value.

```python
d
```
